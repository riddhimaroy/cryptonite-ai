```python
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split

iris = load_iris()
data = iris.data
target = iris.target
feature_names = iris.feature_names
target_names = iris.target_names

print(data[:5])
print (feature_names[::])
print(target[::])
```

```python
X = data
y = target
X_train, X_test, y_train, y_test = train_test_split(X, y)
```

```python
#implementing knn

from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score

knn_classifier = KNeighborsClassifier(n_neighbors=11)

knn_classifier.fit(X_train, y_train)
y_pred = knn_classifier.predict(X_test)

accuracy = accuracy_score(y_test, y_pred)
print("Accuracy:", accuracy)
```

```python
#implementing svm

from sklearn.svm import SVC

svm_classifier = SVC(kernel='linear', C=10.0)

svm_classifier.fit(X_train, y_train)
y_pred = svm_classifier.predict(X_test)

accuracy = accuracy_score(y_test, y_pred)
print("Accuracy:", accuracy)
```

```python
#implementing boosting

from sklearn.ensemble import AdaBoostClassifier

boosting_classifier = AdaBoostClassifier(n_estimators=100, random_state=10)
boosting_classifier.fit(X_train, y_train)
boosting_predictions = boosting_classifier.predict(X_test)
boosting_accuracy = accuracy_score(y_test, boosting_predictions)
print("Accuracy:", boosting_accuracy)
```

```python
#implementing bagging 

from sklearn.ensemble import RandomForestClassifier

bagging_classifier = RandomForestClassifier(n_estimators=100, random_state=10)
bagging_classifier.fit(X_train, y_train)
bagging_predictions = bagging_classifier.predict(X_test)
bagging_accuracy = accuracy_score(y_test, bagging_predictions)
print("Accuracy:", bagging_accuracy)
```
